# Getting data from API
Once we have the API credentials, we can build a function that generates a URL to access the data we're interested in.

```python
# Define functions to create url for API data retrieval

# By state, county, and site
def make_url_bySite(param, bdate, edate, state, county, site):
    base = f"https://aqs.epa.gov/data/api/sampleData/bySite?email={email}&key={key}"
    url = f"{base}&param={param}&bdate={bdate}&edate={edate}&state={state}&county={county}&site={site}"
    return url

# By state
def make_url_byState(param, bdate, edate, state):
    base = f"https://aqs.epa.gov/data/api/sampleData/byState?email={email}&key={key}"
    url2 = f"{base}&param={param}&bdate={bdate}&edate={edate}&state={state}"
    return url2
```

```python
# Define inputs for your function
param = 43502, 43503 # Formaldehyde # parameter code(s) # 43503
bdate = 20240101 # beginning date
edate = 20241231 # end date
state = 40 # state code; OK = 40
county = 109 # county(s) of interest
site = 1037 # site(s) of interest

# Generate urls
site_url = make_url_bySite(param, bdate, edate, state, county, site) # plug values in
print(site_url) # see url

state_url = make_url_byState(param, bdate, edate, state) # plug values in
print(state_url) # see url2
```

```python
# use requests.get function to retreive contents (i.e. data) from a given url
r = requests.get(state_url, verify=False)
```

```python
# use json.loads function to tell the code that the content we want is in json format
contents = json.loads(r.text)

# see the contents of the web page
contents
```

```python
# index by key to only get the contents in "Data" section
data = contents["Data"]
data
```

```python
# put data into DataFrame
data_df = pd.DataFrame(data)  
data_df

# print(data_df['site_number'].unique())
```
